```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
case1 = pd.read_pickle("00_pricing_constant_operation_constant_fi_fit_ne_volumetric_gridch_False_HPnew.pkl")
case2 = pd.read_pickle("00_pricing_dynamic_operation_dynamic_fi_fit_ne_volumetric_gridch_False_HPnew.pkl")

```

```python
day = 70

case1_plot = case1.iloc[day*24:(day+1)*24]
case2_plot = case2.iloc[day*24:(day+1)*24]

case2_plot
```

```python

plt.rcParams.update({'font.size': 14})

fig, (ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8) = plt.subplots(8, sharex=True, figsize=(10, 25))
fig.subplots_adjust(hspace=0.2)  # Adjust vertical space between plots to avoid overlap

# Define a helper function to plot with proper labels
def plot_data(plot_data, column, ax, color, label):
    plot_data[column].plot(ax=ax, color=color, label=label)
    if column == "Deutschland/Luxemburg [€/kWh]":
        column = "Germany/Luxembourg day-ahead spot market price [€/kWh]"
    ax.set_title(column)
    ax.legend()

# Plotting data from case1
plot_data(case1_plot, "Deutschland/Luxemburg [€/kWh]", ax1, "navy", "Constant prices, no flexibility")
plot_data(case1_plot, "Household load [kWh]", ax2, "navy", "Constant prices, no flexibility")
plot_data(case1_plot, "PV generation [kWh]", ax3, "navy", "Constant prices, no flexibility")
plot_data(case1_plot, "BESS charge [kWh]", ax4, "navy", "Constant prices, no flexibility")
plot_data(case1_plot, "BESS discharge [kWh]", ax5, "navy", "Constant prices, no flexibility")
plot_data(case1_plot, "SOC [kWh]", ax6, "navy", "Constant prices, no flexibility")
plot_data(case1_plot, "EV charge [kWh]", ax7, "navy", "Constant prices, no flexibility")
plot_data(case1_plot, "Heat pump demand [kWh]", ax8, "navy", "Constant prices, no flexibility")

# Plotting data from case2
plot_data(case2_plot, "Deutschland/Luxemburg [€/kWh]", ax1, "darkgreen", "Dynamic prices, dynamic operation")
plot_data(case2_plot, "Household load [kWh]", ax2, "darkgreen", "Dynamic prices, dynamic operation")
plot_data(case2_plot, "PV generation [kWh]", ax3, "darkgreen", "Dynamic prices, dynamic operation")
plot_data(case2_plot, "BESS charge [kWh]", ax4, "darkgreen", "Dynamic prices, dynamic operation")
plot_data(case2_plot, "BESS discharge [kWh]", ax5, "darkgreen", "Dynamic prices, dynamic operation")
plot_data(case2_plot, "SOC [kWh]", ax6, "darkgreen", "Dynamic prices, dynamic operation")
plot_data(case2_plot, "EV charge [kWh]", ax7, "darkgreen", "Dynamic prices, dynamic operation")
plot_data(case2_plot, "Heat pump demand [kWh]", ax8, "darkgreen", "Dynamic prices, dynamic operation")

for ax in fig.axes:
    ax.tick_params(axis='x', labelrotation=60)  # Apply rotation to all subplots
    
for ax in fig.axes:
    ax.legend(loc='upper left')

fig.show()
```
